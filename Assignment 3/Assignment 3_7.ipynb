Q7: Go to the link - Set Price filter to “Rs. 6649 to Rs. 13099” , Color filter to “Black”, as shown in the below image.

```python
import selenium
import pandas as pd
from selenium import webdriver
```

```python
driver = webdriver.Chrome(r"C:\Users\Lenovo\Desktop\DS\chromedriver.exe")
driver.get('https://www.myntra.com/shoes')
```

```python
price_filter = driver.find_element_by_xpath ('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[5]/ul/li[2]/label/div')
price_filter.click()
```

```python
color_filter = driver.find_element_by_xpath ('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[6]/ul/li[1]/label/div')
color_filter.click()
```

```python
brand = []
description = []
price = []
```

```python
brand_tag = driver.find_elements_by_class_name('product-brand')
for i in brand_tag:
    bran = i.text
    brand.append(bran)
brand
```

```python
desc = driver.find_elements_by_class_name('product-product')
for i in desc:
    des = i.text
    description.append(des)
description
```

```python
pric = driver.find_elements_by_class_name('product-price')
for i in pric:
    pri = i.text
    price.append(pri)
price
```

```python
print(len(brand),len(description),len(price))
```

```python
df = pd.DataFrame({'Brand':brand[:50],'Short Description':description,'Price':price})
df
```
