Q6: Scrape data for first 100 sneakers you find when you visit flipkart.com and search for “sneakers” in the
search field.

```python
import selenium
import pandas as pd
from selenium import webdriver
```

```python
driver = webdriver.Chrome(r'C:\Users\Lenovo\Desktop\DS\chromedriver.exe')
driver.get ('https://www.flipkart.com/')
```

```python
searc_prod = driver.find_element_by_xpath ("//div[@class = '_3OO5Xc']/input")
searc_prod.send_keys ('sneakers')
```

```python
butt = driver.find_element_by_class_name ('L0Z3Pu')
butt.click()
```

```python
brand = []
description = []
price = []
discount = []
```

```python
brand_tag = driver.find_elements_by_class_name ('_2WkVRV')
for i in brand_tag:
    bran = i.text
    brand.append(bran)
brand
```

```python
descrip = driver.find_elements_by_class_name ('IRpwTa')
for i in descrip:
    des = i.text
    description.append(des)
description
```

```python
price_tag = driver.find_elements_by_class_name ('_30jeq3')
for i in price_tag:
    pri = i.text
    price.append(pri)
price
```

```python
disco = driver.find_elements_by_class_name ('_3Ay6Sb')
for i in disco:
    disc = i.text
    discount.append(disc)
discount
```

```python
df = pd.DataFrame ({'Brand nmae':brand,'Product Description':description,'Price':price[:40],'Discount':discount[:40]})
df
```
