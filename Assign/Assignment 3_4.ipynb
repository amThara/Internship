Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
1. Brand
2. Product Description
3. Price

```python
import selenium
import pandas as pd
from selenium import webdriver
```

```python
driver = webdriver.Chrome(r"C:\Users\Lenovo\Desktop\DS\chromedriver.exe")
driver.get('https://www.flipkart.com/')
```

```python
searc_prod = driver.find_element_by_xpath ("//div[@class = '_3OO5Xc']/input")
searc_prod.send_keys ('sunglasses')
```

```python
butt = driver.find_element_by_xpath ("//button[@class = 'L0Z3Pu']")
butt.click()
```

```python
brand = []
description = []
price = []
discount = []
```

```python
brand_tag = driver.find_elements_by_class_name ('_2WkVRV')
for i in brand_tag:
    bran = i.text
    brand.append(bran)
brand[1:]
```

```python
descrip = driver.find_elements_by_class_name ('IRpwTa')
for i in descrip:
    des = i.text
    description.append(des)
description
```

```python
price_tag = driver.find_elements_by_class_name ('_30jeq3')
for i in price_tag:
    pri = i.text
    price.append(pri)
price
```

```python
disco = driver.find_elements_by_class_name ('_3Ay6Sb')
for i in disco:
    disc = i.text
    discount.append(disc)
discount
```

```python
print (len(brand),len(description),len(price),len(discount))
```

```python
df1 = pd.DataFrame ({'Brand nmae':brand,'Product Description':description,'Price':price[:40],'Discount':discount[:40]})
df1
```
