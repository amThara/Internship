10. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

```python
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException

import warnings
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Chrome(r"C:/Users/Lenovo/Desktop/DS/chromedriver.exe")
driver.get ('https://www.hostelworld.com/')
```

```python
search = driver.find_element_by_xpath("/html/body/div[3]/div/div/div[1]/div[1]/div/div[2]/div[4]/div/div[2]/div/div[1]/div/div/div/input")
search.send_keys('London')
clk = driver.find_element_by_xpath('/html/body/div[3]/div/div/div[1]/div[1]/div/div[2]/div[4]/div/div[2]/div/div[1]/div/div/ul/li[2]/div/span')
clk.click()
```

```python
ccl = driver.find_element_by_xpath('/html/body/div[3]/div/div/div[1]/div[1]/div/div[2]/div[4]/div/div[2]/div/div[5]/button')
ccl.click()
```

```python
url = []
for i in driver.find_elements_by_xpath ("//h2[@class='title title-6']/a"):
    url.append(i.get_attribute('href'))
len(url)
```

```python
url
```

```python
#hostel
hats = []
hat = driver.find_elements_by_xpath("//span[@class='description']")
for i in hat:
    ha = i.text
    hats.append(ha)
hats
```

```python
len(hats)
```

```python
#ratings
ratings = []
for i in url:
    driver.get(i)
    time.sleep(2)
    try:
        rati = driver.find_element_by_xpath("//div[@class='score orange big']")
        ratings.append(rati.text)
    except:
        ratings.append('-')
```

```python
#Overall review
overall = []
for i in url:
    driver.get(i)
    time.sleep(2)
    try:
        over = driver.find_element_by_xpath("//div[@class='keyword']")
        overall.append(over.text)
    except:
        overall.append('-')
```

```python
#total reviews
reviews = []
for i in url:
    driver.get(i)
    time.sleep(2)
    try:
        revi = driver.find_element_by_xpath("//div[@class='reviews']")
        reviews.append(revi.text)
    except:
        reviews.append('-')
```

```python
#property description
description = []
for i in url:
    driver.get(i)
    time.sleep(2)
    try:
        desc = driver.find_element_by_xpath("//div[@class='content collapse-content']")
        description.append(desc.text)
    except:
        description.append('-')
```

```python
#facilities
facilities = []
for i in url:
    driver.get(i)
    time.sleep(2)
    try:
        fac = driver.find_element_by_xpath("//ul[@class='facilities']")
        facilities.append(fac.text)
    except:
        facilities.append('-')
```

```python
#private
private = []
for i in url:
    driver.get(i)
    time.sleep(2)
    try:
        priv = driver.find_element_by_xpath("//div[@id='rate-529933-price']")
        private.append(priv.text)
    except:
        private.append('-')
```

```python
#dorms
dorms = []
for i in url:
    driver.get(i)
    time.sleep(2)
    try:
        dorm = driver.find_element_by_xpath("//div[@id = 'rate-487225-price']")
        dorms.append(dorm.text)
    except:
        dorms.append('-')
```

```python
#hostel name
hostel = []
for i in url:
    driver.get(i)
    time.sleep(2)
    try:
        name = driver.find_element_by_xpath("//div[@class='title-2']")
        hostel.append(name.text)
    except:
        hostel.append('-')
```

```python
print(len(hostel),len(hats),len(ratings),len(reviews),len(overall),len(private),len(dorms),len(facilities),len(description))
```

```python
df = pd.DataFrame({'Hostel Name':hostel,'Distance from City Centre':hats,'Ratings':ratings,'Total reviews':reviews,'Overall Reviews':overall,'Private from price':private,'Dorms from price':dorms,'Facilities':facilities,'Property Description':description})
df
```
