2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
brand = []
product = []
price = []
exchange = []
delivery = []
availability = []
url = []
```

```python
#product
prod = driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-4']")
for i in prod:
    pro = i.text
    product.append(pro)
#price
pric = driver.find_elements_by_class_name ("a-price-whole")
for i in pric:
    pri = i.text
    price.append(pri)
```

```python
#delivery
deliv = driver.find_elements_by_xpath("//span[@class='a-text-bold']")
for i in deliv:
    deli = i.text
    delivery.append(deli)
```

```python
#url
for i in driver.find_elements_by_xpath ("//a[@class = 'a-link-normal s-link-style a-text-normal']"):
    url.append(i.get_attribute('href'))
```

```python
#brand
for i in url:
    driver.get(i)
    time.sleep(2)
    try:
        bran = driver.find_element_by_xpath("//div[@class = 'a-section a-spacing-none']/a")
        brand.append(bran.text)
    except:
        brand.append('-')
```

```python
#availaability
for i in url:
    driver.get(i)
    time.sleep(2)
    try:
        avail = driver.find_element_by_xpath("//div[@class='a-section a-spacing-none }']")
        availability.append(avail.text)
    except:
        availability.append('-')
```

```python
#exchnage/return
for i in url:
    driver.get(i)
    time.sleep(2)
    try:
        exch = driver.find_element_by_xpath("//div[@class = 'a-column a-span3 a-text-center icon-container icon-grid-width celwidget']")
        exchange.append(exch.text)
    except:
        exchange.append('-')
```

```python
print(len(product),len(price),len(delivery),len(url),len(brand),len(exchange),len(availability))
```

```python
df = pd.DataFrame({'Brand Name':brand,'Name of the product':product,'Price':price,'Return/Exchange':exchange,'Expected Delivery':delivery,'Availability':availability,'Product URL':url})
df
```
