9. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.

```python
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException

import warnings
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Chrome(r"C:/Users/Lenovo/Desktop/DS/chromedriver.exe")
driver.get ('https://www.youtube.com/')
```

```python
search = driver.find_element_by_xpath("/html/body/ytd-app/div/div/ytd-masthead/div[3]/div[2]/ytd-searchbox/form/div[1]/div[1]/input")
search.send_keys('web scraping using selenium')
```

```python
butt = driver.find_element_by_xpath("/html/body/ytd-app/div/div/ytd-masthead/div[3]/div[2]/ytd-searchbox/button")
butt.click()
```

```python
clk = driver.find_element_by_xpath("/html/body/ytd-app/div/ytd-page-manager/ytd-search/div[1]/ytd-two-column-search-results-renderer/div/ytd-section-list-renderer/div[2]/ytd-item-section-renderer/div[3]/ytd-video-renderer[2]/div[1]/div/div[1]/div/h3/a/yt-formatted-string")
clk.click()
```

```python
comments = []
comment = driver.find_elements_by_xpath("//ytd-expander[@class='style-scope ytd-comment-renderer']")
for i in comment:
    comm = i.text
    comments.append(comm)
comments
```

```python
len(comments)
```

```python
upvote = []
upv = driver.find_elements_by_xpath("//span[@class='style-scope ytd-comment-action-buttons-renderer']")
for i in upv:
    up = i.text
    upvote.append(up)
```

```python
upvote
```

```python
time = []
tim = driver.find_elements_by_xpath("//yt-formatted-string[@class='published-time-text above-comment style-scope ytd-comment-renderer']")
for i in tim:
    ti = i.text
    time.append(ti)
time
```

```python
len(time)
```

```python
df = pd.DataFrame({'Comments':comments,'Comment upvote':upvote,'Comment time':time})
df
```
