4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
driver = webdriver.Chrome(r"C:/Users/Lenovo/Desktop/DS/chromedriver.exe")
driver.get ('https://www.flipkart.com/')
```

```python
search = driver.find_element_by_xpath("/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input")
search.send_keys('pixel 4A')
```

```python
butt = driver.find_element_by_xpath("/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button")
butt.click()
```

```python
brand = []
phone = []
colour = []
ram = []
rom = []
camera = []
second = []
display = []
battery = []
price = []
urls = []
```

```python
pho = driver.find_elements_by_xpath ("//div[@class='_4rR01T']")
for i in pho:
    phone.append(i.text)
phone
```

```python
brand = [i.split(' ')[0] for i in phone]
brand
```

```python
#url
url = []
urls = driver.find_elements_by_xpath ("//a[@class='_1fQZEK']")
for i in urls:
    url.append(i.get_attribute('href'))
url
```

```python
#price
price = []
pric = driver.find_elements_by_xpath ("//div[@class='_30jeq3 _1_WHN1']")
for i in pric:
    pri = i.text
    price.append(pri)
price
```

```python
#display
display = []
displa = driver.find_elements_by_xpath ("//ul[@class='_1xgFaf']/li[2]")
for i in displa:
    dis = i.text
    display.append(dis)
display
```

```python
#cam
camera = []
came = driver.find_elements_by_xpath ("//ul[@class='_1xgFaf']/li[3]")
for i in came:
    cam = i.text
    camera.append(cam)
camera
```

```python
rear=[]
rear  = [i.split('|')[0] for i in camera]
rear
```

```python
#ram,rom
rar = []
rare = driver.find_elements_by_xpath ("//ul[@class='_1xgFaf']/li[1]")
for i in rare:
    ra = i.text
    rar.append(ra)
rar
```

```python
#ram
ram = []
ram = [i.split ('|')[0] for i in rar]
ram
```

```python
#rom
rom = []
rom = [i.split ('|')[1] for i in rar]
rom
```

```python
#battery
battery = []
batte = driver.find_elements_by_xpath ("//ul[@class='_1xgFaf']/li[4]")
for i in batte:
    bat = i.text
    battery.append(bat)
battery
```

```python
print(len(brand),len(phone),len(ram),len(rom),len(rear),len(display),len(battery),len(price),len(urls))
```

```python
df = pd.DataFrame({'Brand Name':brand,'Smart Phone name':phone,'RAM':ram,'Storage (ROM)':rom,'Camera':rear,'Display Size':display,'Battery Capacity':battery,'Price':price,'Product URL':url})
df
```
