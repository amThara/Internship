# EDA and Data Preprocessing

```python
print(Back.BLACK + Fore.WHITE + '  Visualising the categorical data  ')

plt.figure(figsize = (25,250), facecolor = 'white')
plotnumber = 1

for column in data:
    if plotnumber <=100:
        ax =  plt.subplot(50,2,plotnumber)
        sns.countplot(x = column, data = data)
        plt.xlabel(column, fontsize = 20)
    plotnumber+=1
plt.show
```

From the above graph, there is imbalanced data all over the dataframe

```python
print(Back.BLACK + Fore.WHITE + '  Dropping duplicates if any  ')
print(Style.RESET_ALL)

counter = 0
r,c = data.shape

data = data.copy()
data.drop_duplicates(inplace=True)
data.reset_index(drop=True,inplace=True)

if data.shape==(r,c):
    print('\n The dataset doesn\'t have any duplicates')
else:
    print(f'Number of duplicates dropped {r-data.shape[0]}')
```

Though removing duplicates affects the percentage loss of data, it is necessary for the model to not be biased

```python
print(Back.BLACK + Fore.WHITE + '  Encode the data  ')
for i in data.columns:
    if data[i].dtypes =='object':
        data[i] = enc.fit_transform(data[i].values.reshape(-1,1))
data
```

The data is encoded using ordinal encoder

```python
print(Back.BLACK + Fore.WHITE + '  Checking Null values  ')
data.isnull().sum()
```

There is no null values present

```python
print(Back.BLACK + Fore.WHITE + '  Describing the data  ')

data.describe()
```

All the data in the dataframe is encoded except pin code. So, it has to be standardised

```python
print(Back.BLACK + Fore.WHITE + '  Outliers detection  ')

plt.figure(figsize=(15,50))
graph = 1
for column in data:
    if graph<=45:
        ax = plt.subplot(15,3,graph)
        sns.boxplot(data[column],orient = 'V')
        plt.xlabel(column, fontsize=10)
    graph+=1
    plt.show()
```

There are outliers in data which has to be verified

```python
print(Back.BLACK + Fore.WHITE + '  Removing outliers  ')
print(Style.RESET_ALL)

z = np.abs(zscore(data))
data_new = data[(z<3).all(axis=1)]
print('Before removing outlier', data.shape)
print('After removing outlier', data_new.shape)
```

Since there are only 10 rows having outliers, removing it will give a better result

```python
print(Back.BLACK + Fore.WHITE + '  Checking Correlation  ')
corr_mat = data_new.corr()
plt.figure(figsize=[22,12])
sns.heatmap(corr_mat)
plt.title('Correlation Matrix')
plt.show()
```

```python
data_new.corr()
```

# Feature selection

```python
print(Back.BLACK + Fore.WHITE + '  Multi-collinearity with VIF  ')

vif = pd.DataFrame()
vif['vif'] = [variance_inflation_factor (data_scaled,i)for i in range (data_scaled.shape[1])]
vif['Features'] = data.columns
vif
```

Checked the multicollinearity using variance inflation factor

```python
print(Back.BLACK + Fore.WHITE + '  Principal Component Analysis  ')

pca.fit_transform (data_scaled)
plt.figure ()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Principal Components')
plt.ylabel ('Variance covered')
plt.title ('PCA')
plt.show()
```

On analysing the Principal Component Analysis, it is viewed that 25 features would be enough. Since there are 70 features, we can take 30 features
