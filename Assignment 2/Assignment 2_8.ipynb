```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get('https://www.dineout.co.in/chennai-restaurants/welcome-back')
page
```

```python
#Page content
soup=BeautifulSoup(page.content)
soup
```

```python
# Restaurant names:
restaurant1 = []
for i in soup.find_all('div',class_ = 'restnt-info cursor'):
    restaurant1.append(i.text)
restaurant1
```

```python
restaurant = [i.split(',', 1)[0] for i in restaurant1]
restaurant
```

```python
#Cuisine
cuisine1 = []
for i in soup.find_all('div', class_ = 'detail-info'):
    cuisine1.append(i.text)
cuisine1
```

```python
cuisine2 = [i.split('|', 1)[1] for i in cuisine1]
cuisine2
```

```python
cuisine = [i.split(',', 1)[0] for i in cuisine2]
cuisine
```

```python
# Location:
location1 = []
for i in soup.find_all ('div', class_ = "restnt-loc ellipsis"):
    location1.append (i.text)
location1
```

```python
location = [i.split(',', 1)[0] for i in location1]
location
```

```python
# Ratings
rating = []
for i in soup.find_all('div', class_ = 'restnt-rating rating-4'):
    rating.append(i.text)
rating
```

```python
rating.extend([4,4])
print(rating)
```

```python
# Images URL:
imagesurl = []
for i in soup.find_all ('img', class_ = 'no-img'):
    imagesurl.append (i['data-src'])
imagesurl
```

```python
# Checking the length:
print(len(restaurant),len(cuisine),len(location),len(rating),len(imagesurl))
```

```python
df = pd.DataFrame ({'Restaurant title': restaurant, 'Cuisine': cuisine,'Location': location,'Rating': rating,'Image URL':imagesurl})
df
```
