```python
title = []
titl = driver.find_elements_by_xpath("//h2[@class='_3FpCg']")
for i in titl:
    tit = i.text
    title.append(tit)

kms = []
driven = driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[1]")
for i in driven:
    dri = i.text
    kms.append(dri)

owner = []
onw = driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[2]")
for i in onw:
    own = i.text
    owner.append(own)
    
fuel = []
fuele = driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[3]")
for i in fuele:
    fue = i.text
    fuel.append(fue)
    
price = []
pric = driver.find_elements_by_xpath("//div[@class='_7udZZ']")
for i in pric:
    pri = i.text
    price.append(pri)
    
variant = []
var = driver.find_elements_by_xpath("//p[@class='cvakB']")
for i in var:
    vari = i.text
    variant.append(vari)

year = []
year = [i.split(' ')[0] for i in title]

brand = []
brand = [i.split(' ')[1] for i in title]

model = []
model = [i.split(' ')[2] for i in title]

df1 = pd.DataFrame({'Brand':brand,'Model':model,'Manufactured_year':year,'Running_kms':kms,'No_of_owner': owner,'Fuel_type':fuel,'Variant':variant,'Price':price,'Location':'New Delhi'})
```
