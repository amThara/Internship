1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
    - A) Rank
    - B) Name
    - C) Artist
    - D) Upload date
    - E) Views

```python
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException

import warnings
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Chrome(r"C:/Users/Lenovo/Desktop/DS/chromedriver.exe")
driver.get ('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
names = []
namesd = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
for i in namesd:
    nam = i.text
    names.append(nam)
names
```

```python
hit = []
taken = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
for i in taken:
    tak = i.text
    hit.append(tak)
    
art = []
namesd = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
for i in namesd:
    nam = i.text
    art.append(nam)
    
upler = []
namesd = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
for i in namesd:
    nam = i.text
    upler.append(nam)
    
seen = []
namesd = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
for i in namesd:
    nam = i.text
    seen.append(nam)
```

```python
wiki_df = pd.DataFrame({'Rank':names,'Name':hit,'Artist':art,'Uploaded date':upler,'Views':seen})
```
